### Comments

This comments is based on the differences between the (b) part of the above two questions. In other words, this comments deals with the df_1_n and df_2_n dataframes.


1. Since the words whose frequency is less than 1 within each document is dropped it will reduce the number of terms in the document, that is, the values in N column in df_2 are lesser than that of in df_1. Furthermore, this will increase the TFIDF of all the terms in the document since the N is inversely proprtional to the TF-IDF as shown below:
            
            TF-IDF = ( n/N )*(IDF)

2. The main motive of TF-IDF is to increase the importance to the least occuring terms and lower the importance of frequently occuring words. When words that occur once is removed from consideration, all the frequent terms like a, the, of, etc., showed up in the top 15 per each document as seen in df_2_n when such words rarely made into this list(df_1_n).
